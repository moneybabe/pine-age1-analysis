```python
import numpy as np
from game import sim_game
```

```python
sim_trials = np.empty(10000)
player_bets = np.empty((10000, 12))
other_bets = np.empty((10000, 12))
for i in range(10000):
    player_bets[i], other_bets[i], sim_trials[i] = sim_game()

print(player_bets)
print(sim_trials)
print("Mean player bet: " + str(player_bets.mean()))
print("Mean profit: " + str(sim_trials.mean()))
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize =(10, 7))
ax.hist(sim_trials)
plt.show()
```
